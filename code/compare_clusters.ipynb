```python
import igraph as ig
import networkx as nx
import csv
from collections import defaultdict
```

```python
pre_fname='../data/RectumMicrobiome_PrePost/stool_network_pre.gml.txt.network'
post_fname='../data/RectumMicrobiome_PrePost/stool_network_post.gml.txt.network'
map_fname='../data/raw/shedd.collapsed.norare.map.txt'
raw_fname='../data/raw/shedd.collapsed.norare.biom.txt'

#import from GML format -- generated by WGCNA

G_pre = nx.read_gml(pre_fname, label='id')
print(nx.info(G_pre))
G_post = nx.read_gml(post_fname, label='id')
print(nx.info(G_post))
with open(raw_fname) as f:
    next(f)
    raw_data = defaultdict(dict, [(row['#OTU ID'], row) for row in csv.DictReader(f, delimiter='\t')])
with open(map_fname) as f:  
    map_data = defaultdict(dict, [(row['#SampleID'], row) for row in csv.DictReader(f, delimiter='\t')])
```

```python
pre_fname='../data/1125/Rectum_SparCC/rectum_pre_out_edges.txt'
post_fname='../data/1125/Rectum_SparCC/rectum_post_out_edges.txt'
map_fname='../data/1125/shedd.IL.map.core.square'
raw_fname='../data/1125/shedd.IL.core.0.0001.prob.txt'

# import from edgelists -- generated by SparCC

def import_from_edgelist(fname):
    with open(fname, 'r') as f:
        next(f)
        return nx.parse_edgelist(f.readlines(), nodetype=str, delimiter='\t',data=(('corr',float),('pval',float)))
G_pre = import_from_edgelist(pre_fname)
print(nx.info(G_pre))
G_post = import_from_edgelist(post_fname)
print(nx.info(G_post))
```

```python
# Only works for WGCNA data

def get_attributes(G, attr):
    return set(G.nodes[x][attr] for x in  G.nodes)

pre_OTUID = get_attributes(G_pre, 'OTUID')
post_OTUID = get_attributes(G_post, 'OTUID')
```

```python
# Only works for SparCC data

pre_OTUID = set(G_pre.nodes)
post_OTUID = set(G_post.nodes)
```

```python
print(len(pre_OTUID))
print(len(post_OTUID))
print(len(post_OTUID &  pre_OTUID))
print(len(pre_OTUID - post_OTUID))
print(len(post_OTUID - pre_OTUID))
only_post_biom = post_OTUID - pre_OTUID
only_pre_biom = pre_OTUID - post_OTUID
```

```python
# Only works for WGCNA data

dates = []
for otuid in only_post_biom:
    for k, v in raw_data[otuid].items():
        if v == '0.0':
            continue
        try:
            if map_data[k]['Source2'] == 'dolphin-rectum':
                dates.append(int(map_data[k]['TimeSequence']))
        except (KeyError, ValueError):
            print(k,v)
    break
print(sorted(dates))
```
